```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
myPath = ''
img = cv.imread(myPath+'img/mandrill.tiff',1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB) # Matplotlib espera RGB
plt.imshow(img)

```

```python
# Acceso a pixel
px = img [20,180]
print(px) # B, G, R
```

```python
# Acceso a ROI
roi = img[30:80,150:200]
plt.imshow(roi)
```

```python
# Split and merge channels
r,g,b = cv.split(roi)
plt.imshow(r,cmap='gray')
```

```python
b[:,:]=0 # Hace 0 todos los pixeles azules
img_r = cv.merge((r,g,b))
plt.imshow(img_r)
```
